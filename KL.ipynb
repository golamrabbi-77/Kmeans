```python
import sklearn.cluster as cluster
```

```python
kmeans=cluster.KMeans(n_clusters=3 ,init='k-means++')
kmeans = kmeans.fit(df[['Date','Case']])
```

```python
kmeans.cluster_centers_
```

```python
df.head()
```

```python
df['Clusters']=kmeans.labels_
```

```python
df.head()
```

```python
sns.scatterplot(x="Date", y="Case",hue='Clusters',data=df)
```
